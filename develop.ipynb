```python
from gse_downloader import GSEDownloader


gse_list = ["GSE285498", "GSE179661", "GSE151512", "GSE223827"]

base_dest_dir = "DATA/"

all_archivos = {}
all_raw_counts = {}

for gse_id in gse_list:
    dest_dir = f"{base_dest_dir}{gse_id}/"
    print(f"\nâ¬‡ï¸ Descargando {gse_id} en {dest_dir}")
    
    downloader = GSEDownloader(gse_id, dest_dir)
    archivos, raw_counts = downloader.download()
    
    all_archivos[gse_id] = archivos
    all_raw_counts[gse_id] = raw_counts

for gse_id in gse_list:
    print(f"\nğŸ“‚ {gse_id} - Archivos descargados: {len(all_archivos[gse_id])}")
    if all_raw_counts[gse_id] is not None:
        print(f"ğŸ“Š {gse_id} - Matriz cruda cargada con shape: {all_raw_counts[gse_id].shape}")
    else:
        print(f"âš ï¸ {gse_id} - No se encontrÃ³ matriz cruda")
```

```python
from gse_downloader import GSEDownloader
import pandas as pd

# Ejemplo con un microarray (Series Matrix)
gse_id = "GSE57978"  # Puedes cambiar a cualquier microarray
dest_dir = f"DATA/{gse_id}/"

# Descargar archivos
downloader = GSEDownloader(gse_id, dest_dir)
archivos, _ = downloader.download()  # Para microarray no necesitamos crudos

# Buscar el Series Matrix File
matrix_file = None
for f in archivos:
    if "_series_matrix" in f:
        matrix_file = f
        break

if matrix_file:
    print(f"ğŸ“„ Leyendo Series Matrix File: {matrix_file}")
    # Leerlo como DataFrame
    df = pd.read_csv(matrix_file, sep="\t", comment='!', index_col=0)
    print(f"âœ… Shape del microarray: {df.shape}")
    print(df.head())
else:
    print("âš ï¸ No se encontrÃ³ Series Matrix File para este GSE")
```
